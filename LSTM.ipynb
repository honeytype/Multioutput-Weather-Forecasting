#Modelling

## Create Function

```python
def create_sliding_window(data, target_cols, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size][target_cols])
    return np.array(X), np.array(y)
```

```python
def split_data(X, y, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    total_len = len(X)

    # Validasi total rasio harus 1.0
    assert abs(train_ratio + val_ratio + test_ratio - 1.0) < 1e-6, "Total rasio harus sama dengan 1.0"

    train_end = int(total_len * train_ratio)
    val_end = train_end + int(total_len * val_ratio)

    X_train, y_train = X[:train_end], y[:train_end]
    X_val, y_val = X[train_end:val_end], y[train_end:val_end]
    X_test, y_test = X[val_end:], y[val_end:]

    return (X_train, y_train), (X_val, y_val), (X_test, y_test)
```

## LSTM

### Splitting Data

```python
# Inisialisasi Fitur untuk GRU
features_lstm = ['Tn', 'Tx', 'RR', 'ff_x']
df_lstm = df[features_lstm].copy()  # Buat salinan agar tidak memengaruhi df asli

scaler_lstm = MinMaxScaler()
scaled_data_lstm = scaler_lstm.fit_transform(df_lstm)

# Konversi ke DataFrame lagi agar bisa indexing kolom
scaled_df_lstm = pd.DataFrame(scaled_data_lstm, columns=features_lstm)

# Gunakan sliding window
window_size_lstm = 4  # sama dengan model sebelumnya
target_cols = ['Tn', 'Tx', 'RR', 'ff_x']  # atau kolom target yang kamu mau prediksi
X_lstm, y_lstm = create_sliding_window(scaled_df_lstm.values, scaled_df_lstm.columns.get_indexer(target_cols), window_size_lstm)
```

```python
(X_train_lstm, y_train_lstm), (X_val_lstm, y_val_lstm), (X_test_lstm, y_test_lstm) = split_data(
    X_lstm, y_lstm, train_ratio=0.8, val_ratio=0.1, test_ratio=0.1
)
```

```python
print("X_train_lstm:", X_train_lstm.shape)
print("X_val_lstm:", X_val_lstm.shape)
print("X_test_lstm:", X_test_lstm.shape)
```

### Data Loader

```python
from torch.utils.data import DataLoader, TensorDataset
# Konversi ke tensor untuk lstm
X_train_tensor_lstm = torch.tensor(X_train_lstm, dtype=torch.float32)
y_train_tensor_lstm = torch.tensor(y_train_lstm, dtype=torch.float32)
X_val_tensor_lstm = torch.tensor(X_val_lstm, dtype=torch.float32)
y_val_tensor_lstm = torch.tensor(y_val_lstm, dtype=torch.float32)

# Buat DataLoader untuk lstm
train_loader_lstm = DataLoader(TensorDataset(X_train_tensor_lstm, y_train_tensor_lstm), batch_size=16, shuffle=True)
val_loader_lstm = DataLoader(TensorDataset(X_val_tensor_lstm, y_val_tensor_lstm), batch_size=16)
```

### Define Model

```python
import torch
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=1, dropout=0.3):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_dim, output_dim)  # hanya hidden terakhir

    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]        # Ambil output timestep terakhir
        out = self.fc(out)
        return out
```

```python
# Inisialisasi parameter model lstm
input_dim = X_train_lstm.shape[2]
hidden_dim = 128
output_dim = y_train_lstm.shape[1]  # jumlah target
criterion_lstm = nn.MSELoss()
model_lstm = LSTMModel(input_dim, hidden_dim, output_dim)
optimizer_lstm = torch.optim.Adam(model_lstm.parameters(), lr=0.0001)

num_epochs_lstm = 200
patience_lstm = 20
best_val_loss_lstm = np.inf
wait_lstm = 0
```

### Train Model

```python
train_losses_lstm = []
val_losses_lstm = []

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_lstm.to(device)

for epoch in range(num_epochs_lstm):
    model_lstm.train()
    train_loss = 0.0
    for xb, yb in train_loader_lstm:
        xb, yb = xb.to(device), yb.to(device)
        optimizer_lstm.zero_grad()
        pred = model_lstm(xb)   # Model hanya memerlukan output timestep terakhir
        loss = criterion_lstm(pred, yb)
        loss.backward()
        optimizer_lstm.step()
        train_loss += loss.item()

    model_lstm.eval()
    val_loss = 0.0
    with torch.no_grad():
        for xb, yb in val_loader_lstm:
            xb, yb = xb.to(device), yb.to(device)
            pred = model_lstm(xb)
            loss = criterion_lstm(pred, yb)
            val_loss += loss.item()

    # Rata-rata loss
    avg_train_loss = train_loss / len(train_loader_lstm)
    avg_val_loss = val_loss / len(val_loader_lstm)
    train_losses_lstm.append(avg_train_loss)
    val_losses_lstm.append(avg_val_loss)

    print(f"[lstm] Epoch {epoch+1}/{num_epochs_lstm} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

    # Early Stopping
    if avg_val_loss < best_val_loss_lstm:
        best_val_loss_lstm = avg_val_loss
        wait_lstm = 0
        torch.save(model_lstm.state_dict(), "best_model_lstm.pt")
        print("✅ [LSTM] Best model disimpan.")
    else:
        wait_lstm += 1
        if wait_lstm >= patience_lstm:
            print("⛔️ [LSTM] Early stopping triggered.")
            break
```

```python
# Load the best model for lstm only
model_lstm.load_state_dict(torch.load("best_model_lstm.pt"))
model_lstm.eval()  # Set the model to evaluation mode

# Ubah X_test_lstm menjadi tensor
X_test_lstm_tensor = torch.tensor(X_test_lstm, dtype=torch.float32)

# Lakukan prediksi tanpa gradient tracking
with torch.no_grad():
    y_pred_lstm_tensor = model_lstm(X_test_lstm_tensor)  # Memanggil model lstm untuk prediksi
    y_pred_lstm = y_pred_lstm_tensor.numpy()  # Konversi hasil prediksi ke numpy untuk evaluasi
```

### Evaluation Model

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    epsilon = 1e-8  # untuk mencegah pembagian nol
    return np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100

# Ambil indeks target berdasarkan nama kolom scaler
target_indices = [np.where(scaler_lstm.feature_names_in_ == col)[0][0] for col in target_cols]

# Invers transform y_test
dummy_test = np.zeros((len(y_test_lstm), scaler_lstm.n_features_in_))
for idx, col_idx in enumerate(target_indices):
    dummy_test[:, col_idx] = y_test_lstm[:, idx]
y_test_lstm_inv = scaler_lstm.inverse_transform(dummy_test)[:, target_indices]

# Invers transform prediksi
dummy_pred = np.zeros((len(y_pred_lstm), scaler_lstm.n_features_in_))
for idx, col_idx in enumerate(target_indices):
    dummy_pred[:, col_idx] = y_pred_lstm[:, idx]
y_pred_lstm_inv = scaler_lstm.inverse_transform(dummy_pred)[:, target_indices]

# Inisialisasi dictionary untuk lstm
mse_lstm_dict = {}
rmse_lstm_dict = {}
mae_lstm_dict = {}
mape_lstm_dict = {}

print("📊 Evaluasi lstm pada Test Set:")
for i, var in enumerate(target_cols):
    mse_val = mean_squared_error(y_test_lstm_inv[:, i], y_pred_lstm_inv[:, i])
    rmse_val = np.sqrt(mse_val)
    mae_val = mean_absolute_error(y_test_lstm_inv[:, i], y_pred_lstm_inv[:, i])
    mape_val = mean_absolute_percentage_error(y_test_lstm_inv[:, i], y_pred_lstm_inv[:, i])

    mse_lstm_dict[var] = mse_val
    rmse_lstm_dict[var] = rmse_val
    mae_lstm_dict[var] = mae_val
    mape_lstm_dict[var] = mape_val

    print(f"{var} - MSE: {mse_val:.2f}, RMSE: {rmse_val:.2f}, MAE: {mae_val:.2f}, MAPE: {mape_val:.2f}%")
```
